```python
import cv2
print("package imported")
```

```python
# Read the image
img = cv2.imread("image_path.jpg")

# Convert to grayscale
imggray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Canny edge detection
imgcanny = cv2.Canny(imggray, 150, 200)

# Apply threshold to create binary image (unpack the return tuple)
ret, imgbinary = cv2.threshold(imgcanny, 127, 255, cv2.THRESH_BINARY)

# Display binary image
cv2.imshow("Binary Image", imgbinary)
cv2.waitKey(0)
```
